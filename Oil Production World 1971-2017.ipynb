```python
import pandas as pd
oil_production = pd.read_csv('Crude Oil Production 1971-2017.csv', header = 0)
oil_production.head()
```

```python
oil_production.LOCATION.unique()
```

```python
oil_production_world = oil_production[oil_production.LOCATION == 'WLD']
oil_production_world.LOCATION.unique()
```

```python
cols_names = ['LOCATION', 'INDICATOR', 'SUBJECT', 'MEASURE', 'FREQUENCY', 'Flag Codes']
oil_prod_wld = oil_production_world.drop(columns=cols_names)
```

```python
oil_prod_wld.set_axis(range(len(oil_prod_wld.TIME)), axis='index')
```

```python
oil_prod_wld
```

```python
oil_prod_wld.to_csv('Oil Production World.csv')
```

```python
def growth_rate(serie):
    row = 1
    for row in range(len(serie)-2):
        previous = row -1
        actual = row
        growth = pd.Series(((actual-previous)/previous)*100)
    return growth
        
```

```python
growth_rate(oil_prod_wld.Value, index=False)
```
